```python
# matrix * matrix

from sklearn.datasets import load_iris

iris = load_iris()

x0 = iris.data[0,:]
x1 = iris.data[1,:]
x2 = iris.data[2,:]
x3 = iris.data[3,:]
x4 = iris.data[4,:]

m1 = np.array([x0,x1])
m2 = np.array([x2, x3, x4])

# 2x4 * 4x3 = 2x3
print(m1 @ m2.T)
print(np.dot(m1,m2.T))
```

```python
# law of commutative of matrix
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
c = np.array([[9,8],[7,6]])

# this law is not established
print(a @ b)
print(b @ a)

# law of distribution - established
print(a @ (b + c))
print(a @ b + a @ c)

# these law in transpose operation
print((a + b).T)
print(a.T + b.T)

print((a @ b).T)
print(b.T @ a.T)
```
